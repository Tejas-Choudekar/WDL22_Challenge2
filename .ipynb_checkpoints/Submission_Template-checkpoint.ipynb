### Approach and Methodology

Optimization of public transport in every city is without a question a hot topic. The perfomance of the public transport has direct consequences in the quality of life of the citizens of each city (because of connectivity), and also indirect consequences regarding the overall CO2 emissions of an entire country. It is a big challenge to focus on. Nonetheless, to solve a [Vehicle Routing Problem (VRP)](https://en.wikipedia.org/wiki/Vehicle_routing_problem) of the size of an entire city is a **NP-Hard problem**. This means: Easy to check once it is solved, but **very difficult and time consuming to solve it to optimality**. Because of this, and even before considering to go with Heuristics and other methodologies, our approach and methodology was based on ***Divide and Conquer***. Our idea aims to go from general to specific, and the methodology is shown below


**<center>Figure 1. Proposed Methodology</center>**

From general to specific, the methodology follows the following:

1. **Caiscais Transit Data**:
    - Gather and interpret geospatial data from Caiscais Municipality
    - The data included: 
        - MNIVP Dataset: Dataset with public interventions between 2014 and 2021, with a total of 6617 registers
        - EIXODEVIA Dataset: Dataset with the streets of the city in coordinates (Longitude, Latitude)
        - MNCAUTOCARRO Dataset: Bus routes in coordinates (Longitude, Latitude)
        - GTFS Datasets: Bus routes and stops in coordinates (Longitude, Latitude)
        - POI Data: From Google Maps (Manually) POI data was obtained to check for important areas of the city.
2. **Analysis of Street Interventions**:
    - From the Caiscais data it was obtained the total number of days each street was off service (at least at one point), total number of interrumptions per street, and other type of description of the phenomenon.
    - Based on this information, one street was chosen based on its perfomance (number of interruptions) and its importance to the city (location).
    - The neighborhood of the chosen city is exported to do a route optimization in case of interruption.
3. **Route Optimization on a Single Street**:
    - An Optimization Model for Shortest Path using Dijsktra Exact Algorithm is implement in this part, using the neighborhood selected in part (2) as Network, and the bus route as the subject of optimization.
    - The idea here, is that if it is working for once bus, the methodology is easy to scale to all the buses in the city, in which will be able to include more constraints and to solve the problem using Heuristics such as the Orienteering Algorithm.
