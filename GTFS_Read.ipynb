```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import shapely.geometry
import geopandas as gpd
import contextily as cx

import gtfsutils
import gtfsutils.routes
import gtfsutils.filter
```

```python
%%time
df_dict = gtfsutils.load_gtfs("graph_data/cascais.zip")
df_dict.keys()
```

```python
gdf_routes = gtfsutils.routes.load_routes_counts(df_dict)
gdf_routes.head(2)
```

```python
fig, ax = plt.subplots(figsize=(15, 15))

gdf_plot = gdf_routes.to_crs(epsg=3857)
gdf_plot.plot(ax=ax, column='counts', cmap='OrRd', linewidth=1.2, alpha=0.5, legend=True)

cx.add_basemap(ax, crs=gdf_plot.crs.to_string())
ax.set_title("GTFS Route Counts")
ax.axis('off');
```
